```python
cont_zip=[]
for i in url_zip['0']:
    temp=i+'/graphs/contributors'
    cont_zip.append(temp)
```

```python
conts=[]

for url in tqdm(cont_zip):
    driver.get(url)
    time.sleep(3)
    html = driver.page_source
    soup=BeautifulSoup(html,'html.parser')
    try:
        cont_len=len(soup.find_all('li',{'class':'contrib-person float-left col-6 my-2 pr-2'}))
        cont_name=[]
        cont_commit=[]
        cont_add=[]
        cont_delete=[]
        if cont_len>=5:
            for i in range(5):
                name=driver.find_element_by_xpath('//*[@id="contributors"]/ol/li['+str(i+1)+']/span/h3/a[2]').text
                cont_name.append(name)
                commit=driver.find_element_by_xpath('//*[@id="contributors"]/ol/li['+str(i+1)+']/span/h3/span[2]/span/div/a').text.replace(' commits','')
                cont_commit.append(commit)
                add=driver.find_element_by_xpath('//*[@id="contributors"]/ol/li['+str(i+1)+']/span/h3/span[2]/span/div/span[1]').text.replace(' ++','')
                cont_add.append(add)
                delete=driver.find_element_by_xpath('//*[@id="contributors"]/ol/li['+str(i+1)+']/span/h3/span[2]/span/div/span[2]').text.replace(' --','')
                cont_delete.append(delete)
            conts.append([cont_name,cont_commit,cont_add,cont_delete])

        else:
            for i in range(cont_len):
                name=driver.find_element_by_xpath('//*[@id="contributors"]/ol/li['+str(i+1)+']/span/h3/a[2]').text
                cont_name.append(name)
                commit=driver.find_element_by_xpath('//*[@id="contributors"]/ol/li['+str(i+1)+']/span/h3/span[2]/span/div/a').text.replace(' commits','')
                cont_commit.append(commit)
                add=driver.find_element_by_xpath('//*[@id="contributors"]/ol/li['+str(i+1)+']/span/h3/span[2]/span/div/span[1]').text.replace(' ++','')
                cont_add.append(add)
                delete=driver.find_element_by_xpath('//*[@id="contributors"]/ol/li['+str(i+1)+']/span/h3/span[2]/span/div/span[2]').text.replace(' --','')
                cont_delete.append(delete)
            conts.append([cont_name,cont_commit,cont_add,cont_delete])
    except:
        pass
```
