```python
import pandas as pd
import requests
from selenium import webdriver
import time
import random
from tqdm.notebook import tqdm
import requests
#pip install request
#pip install beautifulsoup4
from bs4 import BeautifulSoup
```

```python
url_zip=pd.read_csv('git_url_file.csv')
```

```python
from selenium import webdriver
driver = webdriver.Chrome('chromedriver')
```

```python
tag_list=[]
readme_list=[]
num_language_list=[]
tag_num_list=[]
cont_num_list=[]
```

```python
for url in tqdm(url_zip['0']):
    driver.get(url)
    html = driver.page_source
    soup=BeautifulSoup(html,'html.parser')
    
    time.sleep(2)

    # 태그 수
    try:
        soup.find('div',{'class':'mt-3 d-flex flex-items-center'})
        tag_l=[]
        num_tag=len(soup.find_all('a',{'class':'topic-tag topic-tag-link'}))
        for i in range(1,num_tag+1):
            tag=driver.find_element_by_xpath('//*[@id="repo-content-pjax-container"]/div/div[2]/div[2]/div/div[1]/div/div[2]/div/a['+str(i)+']')
            tag_l.append(tag.text)
        tag_list.append(tag_l)
        tag_num_list.append(num_tag)
    except:
        tag_l=[]
        num_tag=len(soup.find_all('a',{'class':'topic-tag topic-tag-link '}))
        for i in range(1,num_tag+1):
            tag=driver.find_element_by_xpath('//*[@id="repo-content-pjax-container"]/div/div[2]/div[2]/div/div[1]/div/div[1]/div/a['+str(i)+']')
            tag_l.append(tag.text)
        tag_list.append(tag_l)
        tag_num_list.append(num_tag)
        
        
        
        # readme 
        try:
            readme=soup.find('article',{'class':'markdown-body entry-content container-lg'}).text.replace('\n',' ')
            readme_list.append(readme)
        except:
            readme_list.append('')
        
        # language 수
        try:
            numlang=len(soup.find_all('li',{'class':'d-inline'}))
            num_language_list.append(numlang)
        except:
            num_language_list.append(0)
            
        
        # contributor 수
        try:
            cont_num=driver.find_element_by_xpath('//*[@id="repo-content-pjax-container"]/div/div[2]/div[2]/div/div[5]/div/h2/a').text.split()[1]
            cont_num_list.append(cont_num)
            
        except:
            cont_num_list.append(0)
```

```python
from pandas import Series
a=Series(data=tag_list)
b=Series(data=readme_list)
c=Series(data=num_language_list)
d=Series(data=tag_num_list)
e=Series(data=cont_num_list)
```

```python
git_repo=pd.concat([a,b,c,d,e],axis=1)
git_repo.columns=['tag','readme','num_language','tag_count','contributor_count']
git_repo
```
