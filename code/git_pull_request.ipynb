```python
import pandas as pd
import requests
from selenium import webdriver
import time
import random
from tqdm.notebook import tqdm
import requests
#pip install request
#pip install beautifulsoup4
from bs4 import BeautifulSoup
```

```python
url_zip=pd.read_csv('git_url_file.csv')
url_zip
```

```python
from selenium import webdriver
driver = webdriver.Chrome('chromedriver')
```

```python
pull_zip=[]
for i in url_zip['0']:
    temp=i+'/pulls'
    pull_zip.append(temp)
```

```python
pr_open_list2=[]
pr_closed_list2=[]

for url in tqdm(pull_zip):
    driver.get(url)
    time.sleep(1.5)
    html = driver.page_source
    soup=BeautifulSoup(html,'html.parser')
    
    try:
        open_pr=driver.find_element_by_xpath('//*[@id="js-issues-toolbar"]/div/div[1]/div').text.split()[0]
        pr_open_list2.append(open_pr)
    except:
        pr_open_list2.append('')
    
    
    try:
        closed_pr=driver.find_element_by_xpath('//*[@id="js-issues-toolbar"]/div/div[1]/div').text.split()[2]
        pr_closed_list2.append(closed_pr)
    except:
        pr_closed_list2.append('')
    
```

```python
from pandas import Series
a=Series(data=pr_open_list2)
b=Series(data=pr_closed_list2)
```

```python
git_pr=pd.concat([a,b],axis=1)
git_pr.columns=['pr_open','pr_closed']
git_pr
```
